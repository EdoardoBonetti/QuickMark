```python
import numpy as np
import quickmark.quickmark
help(quickmark.quickmark.ComputeXStar)
```

```python

li = np.random.exponential(scale=1, size= 10**6)
xs = np.linspace(0.0001,1-0.0001,100)
vals = []
for fl in xs: 
    val = quickmark.quickmark.ComputeXStar(li, fl) 
    print(val)
    vals.append(val)
```

```python
import matplotlib.pyplot as plt

plt.plot(xs, vals)
plt.plot(xs, np.sqrt(1-xs))
```
