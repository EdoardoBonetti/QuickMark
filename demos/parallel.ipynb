```python
import numpy as np
import quickmark
from concurrent.futures import ProcessPoolExecutor
import matplotlib.pyplot as plt
import time as time
```

```python

li = np.random.uniform( size= 10**6)
xs = np.linspace(0.0001,1-0.0001,100)
```

```python

vals = []
_t = time.time()
for fl in xs: 
    val = quickmark.quickmark.ComputeXStar(li, fl) 
    vals.append(val)

print(time.time() - _t)
```

```python

plt.plot(xs, vals)
plt.plot(xs, np.sqrt(1-xs))
```

```python


# Function to compute the value for each fl
def compute_value(fl):
    return quickmark.quickmark.ComputeXStar(li, fl)

# Use ProcessPoolExecutor to parallelize
vals = []
_t = time.time()

with ProcessPoolExecutor() as executor:
    vals = list(executor.map(compute_value, xs))

print(time.time()- _t)
```

```python

plt.plot(xs, vals)
plt.plot(xs, np.sqrt(1-xs))
```

```python


vals = []
_t = time.time()
for fl in xs: 
    val = np.sort(li, kind="stable")
    vals.append(val)

print(time.time() - _t)
```
