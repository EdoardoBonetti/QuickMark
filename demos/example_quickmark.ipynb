```python
import numpy as np
import quickmark.quickmark
help(quickmark.quickmark.ComputeXStar)
```

```python

li = np.random.uniform( size= 10**6)
xs = np.linspace(0.0001,1-0.0001,100)
vals = []
for fl in xs: 
    val = quickmark.quickmark.ComputeXStar(li, fl) 
    print(val)
    vals.append(val)
```

```python
import matplotlib.pyplot as plt

plt.plot(xs, vals)
plt.plot(xs, np.sqrt(1-xs))
```

```python
plt.plot(xs, np.abs(np.sqrt(1-xs)- vals))
plt.yscale("log")
```

```python
li2 = np.random.uniform(size= 10**2)
xstar = quickmark.quickmark.ComputeXStar(li2,0.3)


plt.plot(li2,"*")
plt.plot(range(len(li2)),(xstar,)*len(li2))

upper = li2>= xstar
under = li2< xstar

print("sum of all", np.sum(li2))
print("sum of under", np.sum(li2[under]))
print("sum of upper", np.sum(li2[upper]))
print( np.sum(li2[upper])/np.sum(li2))


```
