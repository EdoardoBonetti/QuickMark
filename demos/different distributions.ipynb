# QuickMark on different distributions

```python
import numpy as np
import quickmark
from concurrent.futures import ProcessPoolExecutor
import matplotlib.pyplot as plt
import time as time

```

```python
xs = np.linspace(0.0001,1-0.0001,100)
```

```python
# Function to compute the value for each fl
li_unif = np.random.uniform( size= 10**6)

def compute_value_unif(fl):
    return quickmark.quickmark.ComputeXStar(li_unif, fl)

# Use ProcessPoolExecutor to parallelize
vals = []
_t = time.time()
with ProcessPoolExecutor() as executor:
    vals_unif = list(executor.map(compute_value_unif, xs))

print(time.time()- _t)
```

```python
# Function to compute the value for each fl
li_geom_1 = np.random.exponential(scale =10, size = 10**6)

def compute_value_geom_1(fl):
    return quickmark.quickmark.ComputeXStar(li_geom_1 , fl)

# Use ProcessPoolExecutor to parallelize
vals = []
_t = time.time()
with ProcessPoolExecutor() as executor:
    vals_geom_1 = list(executor.map(compute_value_geom_1, xs))

print(time.time()- _t)
```

```python
# Function to compute the value for each fl
li_geom_2 = np.random.exponential(scale =1, size = 10**6)

def compute_value_geomç_2(fl):
    return quickmark.quickmark.ComputeXStar(li_geom_2 , fl)

# Use ProcessPoolExecutor to parallelize
vals = []
_t = time.time()
with ProcessPoolExecutor() as executor:
    vals_geom_2 = list(executor.map(compute_value_geomç_2, xs))

print(time.time()- _t)
```

```python
import matplotlib.pyplot as plt

plt.loglog(xs, vals_geom_1)
plt.loglog(xs, vals_geom_2)
```

```python
li2 = np.random.exponential(size=10**3)
xstar = quickmark.quickmark.ComputeXStar(li2,0.5)


plt.plot(li2,"*")
plt.plot(range(len(li2)),(xstar,)*len(li2))

upper = li2>= xstar
under = li2< xstar

print("sum of all", np.sum(li2))
print("sum of under", np.sum(li2[under]))
print("sum of upper", np.sum(li2[upper]))
print( np.sum(li2[upper])/np.sum(li2))


```
